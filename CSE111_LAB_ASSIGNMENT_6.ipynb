```python
#TASK 1
class Student:
    ID= 0
    def __init__(self, name, department, age, cgpa):
        self.name= name
        self.dept= department
        self.age= age
        self.cgpa= cgpa
        Student.ID+= 1
        
    def get_details(self):
        print("ID:", Student.ID)
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)
        
    def from_String(shelby):
        list_1= shelby.split("-")
        return Student(list_1[0], list_1[1], list_1[2], list_1[3])

                    
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details() 

print("Class variables can only be assigned when a class has been defined.\
Instance variables, on the other hand, can be assigned or changed at any time.")

print("Instance methods need a class instance and can access the instance through self.\
Class methods don't need a class instance.")
```

```python
#TASK 2
class Assassin:
    num= 0
    def __init__(self, name, sucs_rate):
        self.name= name
        self.sucs_rate= str(sucs_rate)+ "%"
        Assassin.num+= 1
        
    def failureRate(name, fail_rate):
        sucs_rate= 100-fail_rate
        return Assassin(name, sucs_rate)
        
    def failurePercentage(name, fail_percent):
        sucs_rate= 100- int(fail_percent[:-1])
        return Assassin(name, sucs_rate)
        
    def  printDetails(self):
        print("Name:", self.name)
        print("Success rate:", self.sucs_rate)
        print("Total number of Assassin:", Assassin.num)
        
        
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()    
```

```python
#TASK 3
class Passenger:
    count= 0
    def __init__(self, name):
        self.name= name
        Passenger.count+= 1
        
    def set_bag_weight(self, weight):
        if weight<= 20:
            self.bus_fare= 450
        elif 21<=weight<=50:
            self.bus_fare= 450+50
        else:
            self.bus_fare= 450+100
        
    def printDetail(self):
        print(f"Name: {self.name}\nBus Fare: {self.bus_fare}")
    
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#TASK 4
class Travel:
    count= 0
    def __init__(self, place_1, place_2):
        self.place_1= place_1
        self.place_2= place_2
        self.time= 1
        Travel.count+= 1
        
    def set_time(self, time):
        self.time= time
        
    def set_source(self, place_1):
        self.place_1= place_1
        
    def set_destination(self, place_2):
        self.place_2= place_2
        
    def display_travel_info(self):
        return f"Source: {self.place_1}\nDestination:{self.place_2}\nFlight Time:{self.time}:00"
        
        
print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#TASK 5
class Employee:
    def __init__(self, name, workingPeriod):
        self.name= name
        self.workingPeriod= workingPeriod
        
    def employeeByJoiningYear(name, joining_year):
        return Employee(name, 2022- joining_year)
    
    def experienceCheck(workingPeriod, gender):
        if workingPeriod<3:
            if gender== "male":                
                return "He is not experienced"
            elif gender== "female":
                return "She is not experienced" 
        else:
            if gender== "male":                
                return "He is experienced"
            elif gender== "female":
                return "She is experienced"    
            
    
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))    
```

```python
#TASK 7
class Cat:
    Number_of_cats= 0
    def __init__(self, color, pos):
        self.color= color
        self.pos= pos
        Cat.Number_of_cats+= 1
        
    def no_parameter():
        return Cat("White", "sitting")
    
    def first_parameter(color):
        return Cat(color, "sitting")
    
    def second_parameter(pos):
        return Cat("White", pos)
    
    def changeColor(self, color):
        self.color= color
        
    def printCat(self):
        print(f"{self.color} cat is {self.pos}")
        
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)    
```

```python
#TASK 8
import math
class Cylinder:
    radius= 5
    height= 18    
    def __init__(self, radius, height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        Cylinder.radius= radius
        Cylinder.height= height        
        print(f"Updated: radius={radius} and height={height}")
    
    @classmethod
    def swap(class_1, height, radius):
        return class_1(radius, height)

    @classmethod
    def changeFormat(class_1, str_1):
        list_1= str_1.split("-")
        return class_1(float(list_1[0]), float(list_1[1]))
    
    @staticmethod
    def area(radius, height):
        Area= 2*math.pi*radius**2+ 2*math.pi*radius*height
        print("Area:", Area)    
        
    @staticmethod
    def volume(radius, height):
        Volume= math.pi*radius**2*height
        print("Volume:", Volume)
        
    
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)  
```

```python
#TASK 10
class SultansDine:
    total_branch= 0
    total_sell= 0
    list_1= []
    def __init__(self, branch):
        self.name= branch
        SultansDine.total_branch+= 1
        
    def sellQuantity(self, num):
        if num<10:
            self.num= num*300
        elif num<20:
            self.num= num*350
        else:
            self.num= num*400
        SultansDine.total_sell+= self.num   
        
    def branchInformation(self):
        print("Branch Name:", self.name)
        print(f"Branch Sell: {self.num} Taka")
        SultansDine.list_1.append([self.name, self.num])

    def details():
        print("Total Number of branch(s):", SultansDine.total_branch)
        print(f"Total Sell: {SultansDine.total_sell} Taka")
        for i in SultansDine.list_1:
            print("Branch Name:", i[0], end= ", ")
            print(f"Branch Sell: {i[1]} Taka")
            print(f"Branch consists of total sell's: {((i[1]/SultansDine.total_sell)*100):.2f}%")
                

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()                  
```
